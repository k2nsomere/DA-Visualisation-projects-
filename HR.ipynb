```python
Travel_count= HR_data_new['BusinessTravel'].value_counts()
Travel_count
```

```python
Travel_count= HR_data_new.groupby('Attrition')['BusinessTravel'].value_counts().plot(kind ='bar',stacked=True)
plt.xlabel('Travel')
plt.title=('Travel with Attrition') 
plt.show()
```

```python
Department_count = HR_data_new['Department'].value_counts()
Department_count 
```

```python
colors = ['#a9f971','#6890F0','#A890F0']
Department_count.plot.bar(color=colors,figsize=(10,6))
plt.show()
```

```python
Department_count = HR_data_new.groupby(['Department','Attrition']).size().unstack()
Department_count.plot(kind='bar',stacked=False )
plt.xlabel('Department')
plt.title=('Grouped Bar Graph with dataframe') 
plt.show()
```

```python
Gender_count= HR_data_new['Gender'].value_counts()
Gender_count
```

```python
plt.figure(figsize=(10,8))
colors = ['steelblue','pink']
Gender_count.plot.pie(colors = colors,autopct = '%1.0f%%',explode =[0.03]*len(Gender_count),shadow=True)
plt.title('Gender Distribution')
plt.show()
```

```python
Gender_count = HR_data_new.groupby(['Gender','Attrition']).size().unstack()
Gender_count.plot(kind='bar',stacked=False )
plt.xlabel('Department')
plt.title=('Grouped Bar Graph with dataframe') 
plt.show()
```
